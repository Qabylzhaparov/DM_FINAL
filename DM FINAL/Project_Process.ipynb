````python
# ============================================
# II. DATA PREPROCESSING & CLEANING
# Goal: Ensure data quality and suitability for modeling
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ============================================
# 1. MISSING DATA HANDLING
# ============================================

print("Conclusion: There are no missing values in the dataset.")

# ============================================
# 2. OUTLIER ANALYSIS
# ============================================

continuous_features = ["Height", "Weight"]

outlier_summary = {}

for col in continuous_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary[col] = len(outliers)

outlier_report = pd.DataFrame.from_dict(
    outlier_summary, orient="index", columns=["Outlier Count"]
)

display(outlier_report)

# --- Optional: Outlier removal (commented) ---
# for col in numeric_cols:
#     Q1 = df[col].quantile(0.25)
#     Q3 = df[col].quantile(0.75)
#     IQR = Q3 - Q1
#     df = df[(df[col] >= Q1 - 1.5 * IQR) & (df[col] <= Q3 + 1.5 * IQR)]

# ============================================
# 3. VARIABLE TRANSFORMATION
# (Lecture #3)
# ============================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Check skewness
skewness = df[numeric_cols].skew().sort_values(ascending=False)
display(skewness)

# Select highly skewed features
skewed_features = skewness[abs(skewness) > 1].index
print("Highly skewed features:", list(skewed_features))

# --- Apply log transformation ---
df_transformed = df.copy()

for col in skewed_features:
    if (df[col] > 0).all():
        df_transformed[col] = np.log1p(df[col])

# ============================================
# 4. DISTRIBUTION COMPARISON
# ============================================

for col in skewed_features:
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    
    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f"Original {col}")
    
    sns.histplot(df_transformed[col], kde=True, ax=axes[1])
    axes[1].set_title(f"Transformed {col}")
    
    plt.show()

# ============================================
# 5. SUMMARY
# ============================================

print("Data Preprocessing Summary:")
print("- No missing values detected")
print("- Outliers analyzed using IQR method")
print("- Skewed variables transformed using log transformation")
````

````python
# =====================================================
# III. FEATURE ENGINEERING & REDUCTION
# Goal: Create better predictors and manage dimensionality
# =====================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_classif
import matplotlib.pyplot as plt
import seaborn as sns

df_fe = df.copy()

# =====================================================
# 1. BINARY ENCODING (Lecture #2)
# =====================================================

binary_cols = [
    "Gender",
    "family_history_with_overweight",
    "FAVC",
    "SMOKE",
    "SCC"
]

binary_maps = {
    "Gender": {"Male": 1, "Female": 0},
    "family_history_with_overweight": {"yes": 1, "no": 0},
    "FAVC": {"yes": 1, "no": 0},
    "SMOKE": {"yes": 1, "no": 0},
    "SCC": {"yes": 1, "no": 0}
}

for col in binary_cols:
    if col in df_fe.columns:
        df_fe[col] = df_fe[col].map(binary_maps[col])

# =====================================================
# 2. FUZZY / ORDINAL ENCODING (Lecture #2)
# =====================================================

fuzzy_maps = {
    "CAEC": {"no": 0.0, "Sometimes": 0.33, "Frequently": 0.67, "Always": 1.0},
    "CALC": {"no": 0.0, "Sometimes": 0.33, "Frequently": 0.67, "Always": 1.0}
}

for col, mapping in fuzzy_maps.items():
    if col in df_fe.columns:
        df_fe[col] = df_fe[col].map(mapping)

# =====================================================
# 3. DISCRETIZATION (Lecture #2)
# Age, Height, Weight with step of 5
# =====================================================

# Height is in meters  convert to centimeters first
if "Height" in df_fe.columns:
    df_fe["Height_cm"] = df_fe["Height"] * 100

def discretize_step_5(series):
    return (series // 5) * 5

for col in ["Age", "Height_cm", "Weight"]:
    if col in df_fe.columns:
        df_fe[f"{col}_cat"] = discretize_step_5(df_fe[col])

# =====================================================
# 4. NOMINAL ENCODING: MTRANS (Lecture #2)
# =====================================================

if "MTRANS" in df_fe.columns:
    df_fe = pd.get_dummies(
        df_fe,
        columns=["MTRANS"],
        prefix="MTRANS",
        drop_first=True
    )

mtrans_cols = [c for c in df_fe.columns if c.startswith("MTRANS_")]
df_fe[mtrans_cols] = df_fe[mtrans_cols].astype(int)


# =====================================================
# 5. CORRELATION ANALYSIS (Lecture #3)
# =====================================================

plt.figure(figsize=(12, 8))
sns.heatmap(
    df_fe.select_dtypes(include=np.number).corr(),
    cmap="coolwarm",
    annot=False
)
plt.title("Correlation Matrix (Numeric Features)")
plt.show()

# =====================================================
# 6. MUTUAL INFORMATION (Lecture #3)
# =====================================================

target = "NObeyesdad"

# Encode target
le = LabelEncoder()
y = le.fit_transform(df_fe[target])

X = df_fe.drop(columns=[target])
X_numeric = X.select_dtypes(include=np.number)

mi_scores = mutual_info_classif(X_numeric, y, random_state=42)

mi_df = (
    pd.DataFrame({
        "Feature": X_numeric.columns,
        "Mutual Information": mi_scores
    })
    .sort_values(by="Mutual Information", ascending=False)
)

display(mi_df)

# =====================================================
# 7. PCA (Lecture #4)
# =====================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

pca = PCA()
pca.fit(X_scaled)

explained_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(explained_variance, marker="o")
plt.axhline(y=0.95, linestyle="--")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA  Explained Variance")
plt.grid(True)
plt.show()

# PCA with 95% variance
pca_95 = PCA(n_components=0.95)
X_pca_95 = pca_95.fit_transform(X_scaled)

print("Original feature count:", X_numeric.shape[1])
print("Reduced feature count (95% variance):", X_pca_95.shape[1])

# =====================================================
# 8. VIEW PROCESSED DATASET
# =====================================================

display(df_fe.head(20))

# =====================================================
# 9. SUMMARY
# =====================================================

print("Feature Engineering & Reduction Summary:")
print("- Binary, ordinal, and nominal categorical features encoded")
print("- Continuous features discretized with step size of 5")
print("- Feature relevance evaluated using mutual information")
print("- Dimensionality reduced using PCA")
````

````python
# =====================================================
# IV. MODEL DEVELOPMENT & SELECTION (CLASSIFICATION)
# Goal: Predict NObeyesdad
# =====================================================

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# 1. Define Features and Target (Lecture #5)
target = "NObeyesdad"

X = df_fe.drop(columns=[target])
y = df_fe[target]

# Encode target (multiclass)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
````

````python
# 2. TrainTest Split (Lecture #6)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
````

````python
# 3. Logistic Regression (Multinomial)  Lecture #6
log_reg = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        multi_class="multinomial",
        max_iter=200,
        solver="lbfgs"
    ))
])

log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr, target_names=le.classes_))
````

````python
# 4. k-Nearest Neighbors (Lecture #6)
knn_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", KNeighborsClassifier())
])

knn_params = {
    "model__n_neighbors": [3,5,7,9],
    "model__weights": ["uniform", "distance"]
}

knn_grid = GridSearchCV(knn_pipe, knn_params, cv=5, scoring="accuracy")
knn_grid.fit(X_train, y_train)

y_pred_knn = knn_grid.predict(X_test)

print("kNN Best Params:", knn_grid.best_params_)
print("kNN Accuracy:", accuracy_score(y_test, y_pred_knn))
````

````python
# 5. Decision Tree Classifier (Lecture #6)
dt_params = {
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10]
}

dt_grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    dt_params,
    cv=5,
    scoring="accuracy"
)

dt_grid.fit(X_train, y_train)
y_pred_dt = dt_grid.predict(X_test)

print("Decision Tree Best Params:", dt_grid.best_params_)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
````

````python
# 6. Support Vector Machine (Lecture #4 & #6)
svm_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", SVC())
])

svm_params = {
    "model__C": [0.1, 1, 10],
    "model__kernel": ["linear", "rbf"],
    "model__gamma": ["scale", "auto"]
}

svm_grid = GridSearchCV(
    svm_pipe,
    svm_params,
    cv=5,
    scoring="accuracy"
)

svm_grid.fit(X_train, y_train)
y_pred_svm = svm_grid.predict(X_test)

print("SVM Best Params:", svm_grid.best_params_)
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
````

````python
# 7. Ensemble Learning  Random Forest (Bagging) (Lecture #10)
rf_params = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5]
}

rf_grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    rf_params,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)
y_pred_rf = rf_grid.predict(X_test)

print("Random Forest Best Params:", rf_grid.best_params_)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
````

````python
# 8. Model Comparison Summary
results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "kNN",
        "Decision Tree",
        "SVM",
        "Random Forest"
    ],
    "Test Accuracy": [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_knn),
        accuracy_score(y_test, y_pred_dt),
        accuracy_score(y_test, y_pred_svm),
        accuracy_score(y_test, y_pred_rf)
    ],
    "Train Accuracy": [
        accuracy_score(y_train, log_reg.predict(X_train)),
        accuracy_score(y_train, knn_grid.predict(X_train)),
        accuracy_score(y_train, dt_grid.predict(X_train)),
        accuracy_score(y_train, svm_grid.predict(X_train)),
        accuracy_score(y_train, rf_grid.predict(X_train))
    ]
})

results["Overfitting"] = results["Train Accuracy"] - results["Test Accuracy"]
display(results.sort_values(by="Test Accuracy", ascending=False))
````

````python
# 9. Confusion Matrix (Best Model)
best_pred = y_pred_rf  # usually RF performs best

plt.figure(figsize=(8,6))
sns.heatmap(
    confusion_matrix(y_test, best_pred),
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=le.classes_,
    yticklabels=le.classes_
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix  Best Model")
plt.tight_layout()
plt.show()
````

````python
# =====================================================
# V. MODEL EVALUATION & INTERPRETATION
# Goal: Assess model performance and gain insights
# =====================================================

from scipy import stats
from sklearn.metrics import precision_score, recall_score, f1_score

# =====================================================
# 1. OVERFITTING DETECTION (Lecture #7)
# =====================================================

print("Overfitting Analysis:")
print("=" * 50)
for model_name, model_obj, train_pred, test_pred in [
    ("Logistic Regression", log_reg, log_reg.predict(X_train), y_pred_lr),
    ("kNN", knn_grid, knn_grid.predict(X_train), y_pred_knn),
    ("Decision Tree", dt_grid, dt_grid.predict(X_train), y_pred_dt),
    ("SVM", svm_grid, svm_grid.predict(X_train), y_pred_svm),
    ("Random Forest", rf_grid, rf_grid.predict(X_train), y_pred_rf)
]:
    train_acc = accuracy_score(y_train, train_pred)
    test_acc = accuracy_score(y_test, test_pred)
    diff = train_acc - test_acc
    status = " OVERFITTING" if diff > 0.1 else " OK"
    print(f"{model_name:20s} | Train: {train_acc:.4f} | Test: {test_acc:.4f} | Diff: {diff:.4f} {status}")

# =====================================================
# 2. FEATURE IMPORTANCE (Tree-based models) (Lecture #8)
# =====================================================

# Random Forest Feature Importance
rf_best = rf_grid.best_estimator_  # Random Forest doesn't use pipeline
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_best.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(data=feature_importance.head(15), y='Feature', x='Importance')
plt.title('Top 15 Feature Importance (Random Forest)')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

display(feature_importance.head(10))

# Decision Tree Feature Importance
dt_best = dt_grid.best_estimator_
dt_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': dt_best.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(data=dt_importance.head(15), y='Feature', x='Importance')
plt.title('Top 15 Feature Importance (Decision Tree)')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

# =====================================================
# 3. ANOVA ANALYSIS (Lecture #8)
# =====================================================

# Calculate BMI for ANOVA
df_fe['BMI'] = df_fe['Weight'] / (df_fe['Height'] ** 2)

# Perform ANOVA: Test if BMI differs significantly across NObeyesdad classes
obesity_groups = [df_fe[df_fe['NObeyesdad'] == cat]['BMI'].values 
                  for cat in df_fe['NObeyesdad'].unique()]

f_stat, p_value = stats.f_oneway(*obesity_groups)

print("\nANOVA Analysis (BMI across NObeyesdad classes):")
print("=" * 50)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.2e}")

if p_value < 0.05:
    print(" Significant difference (p < 0.05): BMI differs across obesity classes")
else:
    print(" No significant difference (p >= 0.05)")

# Visualize BMI distribution by class
plt.figure(figsize=(12, 6))
df_fe.boxplot(column='BMI', by='NObeyesdad', ax=plt.gca())
plt.title('BMI Distribution by NObeyesdad Class')
plt.suptitle('')  # Remove default title
plt.xlabel('NObeyesdad Class')
plt.ylabel('BMI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 4. DETAILED PERFORMANCE METRICS
# =====================================================

print("\nDetailed Performance Metrics (Random Forest - Best Model):")
print("=" * 50)
print(classification_report(y_test, y_pred_rf, target_names=le.classes_))

# =====================================================
# SUMMARY
# =====================================================
print("\nModel Evaluation Summary:")
print("- Overfitting detected and analyzed")
print("- Feature importance extracted from tree-based models")
print("- ANOVA analysis performed on BMI across classes")
print("- Detailed performance metrics calculated")
````

````python
# =====================================================
# PREDICTION FUNCTION EXAMPLE
# =====================================================

def predict_obesity_level(input_data):
    """
    Predict obesity level from input data.
    
    Parameters:
    input_data: dict, pandas Series, or pandas DataFrame with feature values
                Must include all features used in training
    
    Returns:
    predicted_class: str, the predicted obesity level
    """
    import pandas as pd
    
    # Load model
    with open("obesity_model.pkl", "rb") as f:
        model_package = pickle.load(f)
    
    model = model_package['model']
    label_encoder = model_package['label_encoder']
    feature_names = model_package['feature_names']
    
    # Convert input to DataFrame with correct feature order
    if isinstance(input_data, pd.DataFrame):
        # Already a DataFrame, use it directly (don't wrap in list!)
        input_df = input_data.copy()
    elif isinstance(input_data, pd.Series):
        # Series: convert to DataFrame (single row)
        input_df = pd.DataFrame([input_data.to_dict()])
    elif isinstance(input_data, dict):
        # Dictionary: convert to DataFrame
        input_df = pd.DataFrame([input_data])
    else:
        # Try to convert to DataFrame (for numpy arrays, lists, etc.)
        try:
            input_df = pd.DataFrame(input_data)
        except:
            # Last resort: wrap in list
            input_df = pd.DataFrame([input_data])
    
    # Ensure all features are present and in correct order
    for feature in feature_names:
        if feature not in input_df.columns:
            input_df[feature] = 0  # Default value if missing
    
    input_df = input_df[feature_names]  # Reorder columns
    
    # Make prediction
    prediction_encoded = model.predict(input_df)
    prediction = label_encoder.inverse_transform(prediction_encoded)
    
    return prediction[0]

# Example usage (using a sample from test set)
print("Example Prediction:")
print("=" * 50)
sample_idx = 0
sample_data = X_test.iloc[sample_idx:sample_idx+1]
predicted = predict_obesity_level(sample_data)
actual = le.inverse_transform([y_test[sample_idx]])[0]
print(f"Predicted: {predicted}")
print(f"Actual: {actual}")
print(f"Match: {'' if predicted == actual else ''}")
````
